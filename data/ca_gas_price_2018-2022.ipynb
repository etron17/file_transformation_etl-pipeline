```python
# import pandas libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
```

```python
# import data frames 
path = 'gas-ca-2018-2022.csv'
```

```python
df = pd.read_csv(path)
```

```python
df
```

```python
# Reshape from Wide to Long
df = pd.melt(df, id_vars = 'Geography', var_name = "Date", value_name = "Price")
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%y-%b')
df['Date'] = df['Date'].dt.strftime('%b-%Y')
```

```python
df
```

```python
distinct_date = df['Date'].unique().tolist()
distinct_date
```

```python
df.info()
```

```python
distinct_geography = df['Geography'].unique().tolist()
```

```python
distinct_geography
```

```python
df[df['Geography'] == 'Toronto, Ontario'].plot(x='Date', y='Price')
```

```python
# separate city and province
df[['City', 'Province']] = df['Geography'].str.extract(r'(.*), (.*)')
```

```python
df
```

```python
df.head(19)
```

```python
# replace city name
df['City'] = df['City'].str.replace('Ottawa-Gatineau, Ontario part', 'Ottawa')
df['Province'] = df['Province'].str.replace('Ontario/Quebec 2', 'Ontario')
df['Province'] = df['Province'].str.replace('Ontario 3', 'Ontario')
df['Province'] = df['Province'].str.replace('Alberta 4', 'Alberta')
```

```python
df.head(50)
```

```python
df = df[df.Geography != 'Canada']
```

```python
df
```

```python
df = df[['City','Province', 'Price', 'Date']]
```

```python
df
```

```python
df.to_csv('avg_gas_price_ca.csv')
```
